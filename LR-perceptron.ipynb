```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                          n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
# np.insert(X,0,1,axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1) # adds a column of 1 in each row 
    weights = np.ones(X.shape[1]) # makes an array of weights like W0,W1,W2
    lr=0.1

    for i in range(1000):
        j = np.random.randint(0,100) # chooses a random value for classification
        y_hat = step(np.dot(X[j],weights)) #calculates the value of WiXi >/< 0
        weights = weights + lr*(y[j]-y_hat)*X[j] #calculates the value of new weights
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0])/(coef_[1])
b = -(intercept_)/(coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b 
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter', s=100)
plt.ylim(-4,2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0])/(lor.coef_[0][1])
b = -(lor.intercept_)/(lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='blue',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
''' So the difference between the perceptron line and the sklearn logistic regression line
Is that the per line stops when all the given points are classified, but the sk lr line gives 
us the most symmetric partition between the negative and positive regions, we can see that by 
increasing the separation between the regions.'''

from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                          n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=15)
```

```python
m = -(lor.coef_[0][0])/(lor.coef_[0][1])
b = -(lor.intercept_)/(lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```
